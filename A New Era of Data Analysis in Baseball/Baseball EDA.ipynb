```python
judge_events_2017 = judge[judge['game_date'].apply(lambda x: x[:4]) == '2017']['events']
print('Aaron Judge stats till 2017')
judge_events_2017.value_counts()
```

```python
stanton_events_2017 = stanton[stanton['game_date'].apply(lambda x: x[:4]) == '2017']['events']
print('Giancarlo Stanton stats till 2017')
stanton_events_2017.value_counts()
```

```python
judge_hr = judge[judge['events'].apply(lambda x: x == 'home_run')]

stanton_hr = stanton[stanton['events'].apply(lambda x: x == 'home_run')]
```

```python
fig1, axs1 = plt.subplots(ncols=2, sharex=True, sharey=True)

sns.regplot(x='launch_speed', y='launch_angle', fit_reg=True, color='tab:blue', data=judge_hr, ax=axs1[0]).set_title('Aaron Judge\nHome Runs, 2015-2017')
sns.regplot(x='launch_speed', y='launch_angle', fit_reg=True, color='tab:blue', data=stanton_hr, ax=axs1[1]).set_title('Giancarlo Stanton\nHome Runs, 2015-2017')
```

```python
fig2, axs2 = plt.subplots(ncols=2, sharex=True, sharey=True)

sns.kdeplot(judge_hr['launch_speed'], judge_hr['launch_angle'], cmap="Blues", shade=True, shade_lowest=False, ax=axs2[0]).set_title('Aaron Judge\nHome Runs, 2015-2017')
sns.kdeplot(stanton_hr['launch_speed'], stanton_hr['launch_angle'], cmap="Blues", shade=True, shade_lowest=False, ax=axs2[1]).set_title('Giancarlo Stanton\nHome Runs, 2015-2017')
```

```python
judge_stanton_hr = pd.concat([judge_hr, stanton_hr])

sns.boxplot(y='release_speed', x='player_name', data=judge_stanton_hr, color='tab:blue').set_title('Home Runs, 2015-2017')
```

```python
def assign_x_cord(row):
    if row.zone in [1, 4, 7]:
        return 1
    elif row.zone in [2, 5, 8]:
        return 2
    elif row.zone in [3, 6, 9]:
        return 3
```

```python
def assign_y_cord(row):
    if row.zone in [1, 2, 3]:
        return 3
    elif row.zone in [4, 5, 6]:
        return 2
    elif row.zone in [7, 8, 9]:
        return 1
```

```python
judge_strike_hr = judge_hr[judge_hr['zone'] <= 9]
judge_strike_hr

judge_strike_hr['zone_x'] = judge_strike_hr.apply(assign_x_cord, axis=1)
judge_strike_hr['zone_y'] = judge_strike_hr.apply(assign_y_cord, axis=1)

plt.hist2d(judge_strike_hr['zone_x'], judge_strike_hr['zone_y'], bins=3, cmap='Blues')
plt.title('Aaron Judge Home Runs on\n Pitches in the Strike Zone, 2015-2017')
plt.gca().get_xaxis().set_visible(False)
plt.gca().get_yaxis().set_visible(False)
cb = plt.colorbar()
cb.set_label('Counts in Bin')
```

```python
stanton_strike_hr = stanton_hr[stanton_hr['zone'] <= 9]
stanton_strike_hr

stanton_strike_hr['zone_x'] = stanton_strike_hr.apply(assign_x_cord, axis=1)
stanton_strike_hr['zone_y'] = stanton_strike_hr.apply(assign_y_cord, axis=1)

plt.hist2d(stanton_strike_hr['zone_x'], stanton_strike_hr['zone_y'], bins=3, cmap='Blues')
plt.title('Giancarlo Stanton Home Runs on\n Pitches in the Strike Zone, 2015-2017')
plt.gca().get_xaxis().set_visible(False)
plt.gca().get_yaxis().set_visible(False)
cb = plt.colorbar()
cb.set_label('Counts in Bin')
```
